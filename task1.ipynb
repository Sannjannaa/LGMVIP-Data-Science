Image to Pencil Sketch with Python

```python
import cv2
from matplotlib import pyplot as plt
```

```python
image=cv2.imread("doraemon.jpg")
plt.imshow(image)
plt.title("Image")
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
plt.title("Grayscale Image")
```

```python
inverted_gray_image=255-gray
plt.imshow(inverted_gray_image,cmap="gray")
plt.title("Inverted Greyscale Image")
```

```python
blurred_image=cv2.GaussianBlur(inverted_gray_image,(21,21),0)
inverted_gray_image=255-gray
plt.imshow(blurred_image,cmap="gray")
plt.title("Blurred Image")
```

```python
pencil_sketch=cv2.divide(gray,inverted_gray_image,scale=256.0)
plt.imshow(pencil_sketch,cmap="gray")
plt.title("Pencil Sketch")
```
